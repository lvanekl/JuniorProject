```python
import matplotlib.pyplot as plt
from skimage import io
import skimage
from skimage.filters import threshold_otsu
import numpy as np
from my_functions import to_gray
import pandas as pd
```

```python
image_name = r'data\images\3155.png'
raw_image = io.imread(image_name)[:,:,:3]
gray_image = to_gray(raw_image)


thresh = threshold_otsu(gray_image)


mask = gray_image > thresh

OTSU_image = raw_image.copy()
OTSU_image[mask] = (0, 0, 0)
plt.imshow(OTSU_image)
```
